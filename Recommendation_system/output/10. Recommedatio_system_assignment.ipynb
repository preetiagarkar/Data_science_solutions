```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity
```

```python
book.drop(['Unnamed: 0'], axis=1, inplace=True)
book = book.rename(columns={'User.ID':'userid' , 'Book.Title':'title' ,'Book.Rating':'rating'} )
book.head()
```

```python
user_book_table = pd.pivot_table(data=book,index='userid', columns='title', values='rating').reset_index(drop=True)
user_book_table
```

```python
user_book_table.fillna(0, inplace=True)
```

```python
user_book_table
```

```python
# Directly finding cosine similarity

user_sim = cosine_similarity(user_book_table.values)
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
user_sim_df.index = book.userid.unique()
user_sim_df.columns = book.userid.unique()
user_sim_df
```
